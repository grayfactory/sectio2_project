```python
import requests
from bs4 import BeautifulSoup
from pathlib import Path
import pandas as pd
from re import search
import numpy as np
from time import sleep
import csv
```

```python
path = Path('.')
files = [p for p in path.glob('*.csv')]
files
```

```python
links = pd.read_csv(files[1], dtype={'imdbId':'str'})
links
```

#### url 규칙 
- {movie_id}자리에 'tt+imdbId'가 들어간다. https://www.imdb.com/title/{movie_id}/?ref_=fn_al_tt_1

#### tag 규칙
- div tag, id=titleDetails 에 해당 정보들이있음
- pin point로 찾을 수 없었기 때문에 line by line으로 돌면서 찾는 text가 있을 때 정보를 get

#### 시간이 오래 걸려서 3개로 나눠서 동시에 진행

```python
with open('info_budget_2.csv','w') as csvfile:
    fieldnames = ['imdbId','budget','gross_usa','gross_world']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    
    for m in links.iloc[15281:30562].imdbId:
        movie_id = 'tt'+m
        print(movie_id)
        url = f'https://www.imdb.com/title/{movie_id}/?ref_=fn_al_tt_1'
        r=requests.get(url)
        sleep(np.round(np.random.random_sample(),3))
        
        soup = BeautifulSoup(r.text, 'html.parser')
        
        for i in soup.select("div#titleDetails"):
            budget, usa, world = [],[],[]
            for j in i.find_all("div"):
                texts = j.text
                if search('Budget',texts):
                    budget = j.text.split(':')[1].replace(' ','')

                if search('Gross USA',texts):
                    usa = j.text.split(':')[1].replace(' ','')

                if search('Worldwide Gross',texts):
                    world = j.text.split(':')[1].replace(' ','')

            
            writer.writerow({'imdbId': movie_id
                              ,'budget': budget
                              ,'gross_usa': usa
                              ,'gross_world': world
                             })
    
```
